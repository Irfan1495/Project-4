```python
# Import Dependencies
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn import tree
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
```

### Looking at Data

```python
# Read the vgsales.csv and convert to dataframe
vgsales_df = pd.read_csv("vgsales.csv")

# View dataframe
vgsales_df.head()
```

```python
# Drop NaN values because they won't be useful in our analysis
vgsales_df = vgsales_df.dropna()
```

```python
# Get info on dataframe
vgsales_df.shape
vgsales_df.info()
vgsales_df.describe()
```

```python
# Convert year column from float to int
vgsales_df["Year"] = vgsales_df["Year"].astype(int)

# View dataframe
vgsales_df.head()
```

### Preprocessing Data

```python
# Create a copy of the dataframe
vgsales_copy_df = vgsales_df.copy()
```

```python
# Drop all columns except Platform, Genre, and NA_Sales
vgsales_copy_df = vgsales_copy_df.drop(columns=['Year', 'Publisher', 'Rank', 'Name', 
                                              'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales'])

# View dataframe
vgsales_copy_df.head()
```

```python
# Bin Platform data to reduce overfit and simplify the model
dev_bins = {
    'sony': ['PS3', 'PS2', 'PS4', 'PS', 'PSP', 'PSV'], 
    'microsoft': ['X360', 'XOne', 'XB', 'PC'], 
    'nintendo': ['Wii', 'GB', 'DS', 'GBA', '3DS', 'GC', 'WiiU', 'NES', 'SNES', 'N64'], 
    'sega': ['GEN', 'DC', 'SAT', 'SCD', 'GG'], 
    'misc': ['NG', 'TG16', '3DO', 'PCFX', '2600', 'WS']
}

# Define a function to assign bins
def assign_bins(platform):
    for dev, consoles in dev_bins.items():
        if platform in consoles:
            return dev

# Create a new column for the assigned bins
vgsales_copy_df['platform_bins'] = vgsales_copy_df['Platform'].apply(assign_bins)

# Drop the original Platform column
vgsales_copy_df = vgsales_copy_df.drop(columns='Platform')

# View dataframe
vgsales_copy_df.head()
```

```python
# Create bins and labels to bin NA_sales
bins = [0, 1, 2, 3, 4, 5, 10, 50]
labels = ['bad', 'very_low', 'low', 'medium_low', 'medium', 'medium_high', 'high']

# Create a new column for the binned data
vgsales_copy_df['sales_category'] = pd.cut(vgsales_copy_df['NA_Sales'], bins=bins, 
                                          labels=labels, right=False)

# View dataframe
vgsales_copy_df.head()
```

```python
# Create a new column for the binned data
vgsales_copy_df['sales_category'] = pd.cut(vgsales_copy_df['NA_Sales'], bins=bins, 
                                          labels=labels, right=False)

# Drop NA_Sales from dataframe
vgsales_copy_df = vgsales_copy_df.drop(columns='NA_Sales')

# View dataframe
vgsales_copy_df.head()
```

```python
# Reset index
vgsales_copy_df.reset_index(drop=True, inplace=True)

#View dataframe
vgsales_copy_df.head()
```

```python
# Use OneHotEncoder to change Genre and platform_bins into numerical binary features
enc = OneHotEncoder()

# Create a dataframe that is fit and trasformed
enc_data = pd.DataFrame(enc.fit_transform(vgsales_copy_df[['Genre', 'platform_bins']]).toarray())

# Change column names to represent features
enc_data.columns = enc.get_feature_names_out()

# View dataframe
enc_data.head()
```

```python
# Create a new dataframe and join enc to vgsales_copy_df
vgsales_enc_df = vgsales_copy_df.join(enc_data)

# View dataframe
vgsales_enc_df.head()
```

### Create and Train a Decision Tree Model

```python
# Define target variable and features
y = vgsales_enc_df['sales_category']
X = vgsales_enc_df.drop(columns=['sales_category', 'Genre', 'platform_bins'])
```

```python
#View features
X.head()
```

```python
# Create train and test data 
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# View train and test sizes
print("train size X : ", X_train.shape)
print("train size y : ", y_train.shape)
print("test size X : ", X_test.shape)
print("test size y : ", y_test.shape)
```

```python
# Create model
model = tree.DecisionTreeClassifier()

# Fit model
model.fit(X_train, y_train)
```

```python
# Make prediction using X_test
y_pred = model.predict(X_test)
```

```python
# View accuracy score
accuracy = accuracy_score(y_test, y_pred)
print('accuracy:', accuracy)
```

```python
# Visualize the model and save as a pdf
plt.figure(figsize=(30,10))
tree.plot_tree(model, fontsize=10, feature_names=list(X.columns), max_depth=6, class_names=labels)
plt.savefig('tree.pdf')
plt.show()
```
