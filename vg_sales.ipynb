```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn import tree
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
```

```python
vgsales_df = pd.read_csv("vgsales.csv")
vgsales_df.head()
```

```python
vgsales_df = vgsales_df.dropna()
```

```python
vgsales_df["Year"] = vgsales_df["Year"].astype(int)
vgsales_df.head()
```

```python
vgsales_df.shape
vgsales_df.info()
vgsales_df.describe()
```

```python
vgsales_copy_df = vgsales_df.copy()
```

```python
vgsales_copy_df = vgsales_copy_df.drop(columns=['Year', 'Publisher', 'Rank', 'Name', 
                                              'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales'])
vgsales_copy_df.head()
```

```python
#vgsales_copy_df['Platform'].unique()
dev_bins = {
    'sony': ['PS3', 'PS2', 'PS4', 'PS', 'PSP', 'PSV'], 
    'microsoft': ['X360', 'XOne', 'XB', 'PC'], 
    'nintendo': ['Wii', 'GB', 'DS', 'GBA', '3DS', 'GC', 'WiiU', 'NES', 'SNES', 'N64'], 
    'sega': ['GEN', 'DC', 'SAT', 'SCD', 'GG'], 
    'misc': ['NG', 'TG16', '3DO', 'PCFX', '2600', 'WS']
}
def assign_bins(platform):
    for dev, consoles in dev_bins.items():
        if platform in consoles:
            return dev
vgsales_copy_df['platform_bins'] = vgsales_copy_df['Platform'].apply(assign_bins)
vgsales_copy_df = vgsales_copy_df.drop(columns='Platform')
vgsales_copy_df
```

```python
bins = [0, 1, 2, 3, 4, 5, 10, 50]
labels = ['bad', 'very_low', 'low', 'medium_low', 'medium', 'medium_high', 'high']
```

```python
vgsales_copy_df['sales_category'] = pd.cut(vgsales_copy_df['NA_Sales'], bins=bins, 
                                          labels=labels, right=False)
vgsales_copy_df
```

```python
#vgsales_copy_df['Genre'] = vgsales_copy_df['Genre'].astype('category')
#vgsales_copy_df['platform_bins'] = vgsales_copy_df['platform_bins'].astype('category')
#vgsales_copy_df['sales_category'] = vgsales_copy_df['sales_category'].astype('category')
```

```python
#vgsales_copy_df['Genre_new'] = vgsales_copy_df['Genre']
#vgsales_copy_df['platform_bins_new'] = vgsales_copy_df['platform_bins']
#vgsales_copy_df['sales_category_new'] = vgsales_copy_df['sales_category']
vgsales_copy_df.reset_index(drop=True, inplace=True)
#vgsales_copy_df
```

```python
enc = OneHotEncoder()
enc_data = pd.DataFrame(enc.fit_transform(vgsales_copy_df[['Genre', 'platform_bins']]).toarray())
```

```python
first_row = vgsales_copy_df[['Genre', 'platform_bins']].iloc[0:1,:]
first_row
```

```python
nin_sport = enc.transform(first_row).toarray()
```

```python
enc_data.columns = enc.get_feature_names_out()
enc_data
```

```python
new_df = vgsales_copy_df.join(enc_data)
new_df
```

```python
y = new_df['sales_category']
X = new_df.drop(columns=['sales_category', 'NA_Sales', 'Genre', 'platform_bins'])
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=50)

print("train size X : ", X_train.shape)
print("train size y : ", y_train.shape)
print("test size X : ", X_test.shape)
print("test size y : ", y_test.shape)
```

```python
#scaler = StandardScaler()
#X_scaler = scaler.fit(X_train)
#X_train_scaled = X_scaler.transform(X_train)
#X_test_scaled = X_scaler.transform(X_test)
```

```python
model = tree.DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
np.unique(y_pred)
```

```python
accuracy = accuracy_score(y_test, y_pred)
#mse = mean_squared_error(y_test, y_pred)
print('accuracy:', accuracy)
#mse
```

```python
plt.figure(figsize=(30,10))
tree.plot_tree(model, fontsize=10, feature_names=list(X.columns), max_depth=6, class_names=labels)
plt.savefig('tree.pdf')
plt.show()
```
